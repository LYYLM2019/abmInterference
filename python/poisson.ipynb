```python
import math
import matplotlib.pyplot as plt

def f(x, lambdaa):
    return ((lambdaa**x) * math.exp(-lambdaa))/math.factorial(x)
```

```python
col = []
lambdaa = [0.5,5,50]
for l in lambdaa:
    plot = []
    for i in range(70):
        plot.append(f(i,l))
    col.append(plot)
```

```python
plt.figure(figsize = (16,8))
plt.plot(col[0], label = ('lambda = %.3f' % lambdaa[0]))
plt.plot(col[1], label = ('lambda = %.3f' % lambdaa[1]))
plt.plot(col[2], label = ('lambda = %.3f' % lambdaa[2]))
plt.legend()
plt.show()
```

What is the probability of at least x < 10 when lambda = 20? __Probability density function?__

```python
def poissonPDF(x,lambdaa):
    get = i = 0
    while i < x:
        get = get + f(i,lambdaa)
        i = i + 1
    return 1 - get
```

```python
col = []
lambdaa = [0.5,5,50]
for l in lambdaa:
    plot = []
    for i in range(50):
        plot.append(poissonPDF(i,l))
    col.append(plot)
```

```python
plt.figure(figsize = (16,8))
plt.plot(col[0], label = ('lambda = %.1f' % lambdaa[0]))
plt.plot(col[1], label = ('lambda = %.1f' % lambdaa[1]))
plt.plot(col[2], label = ('lambda = %.1f' % lambdaa[2]))
plt.legend()
plt.show()
```
